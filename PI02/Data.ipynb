```python
# Importamos las bibliotecas necesarias: pandas para trabajar con datos en forma de tablas y numpy para operaciones numéricas.
import pandas as pd
import numpy as np
```

```python
# Leemos el archivo CSV "AccidentesAviones.csv" y carga los datos en un DataFrame llamado "df".
df = pd.read_csv("AccidentesAviones.csv")
df
```

```python
# Dropeamos la primera columna que es solo los indices duplicados
df.drop(columns=["Unnamed: 0"], inplace=True)
df
```

```python
# como los nombres de las columnas no son claros ni comparten un formato entre si los cambiamos
df.rename(columns={'HORA declarada': "hora_declarada", 'Ruta': "ruta", 'OperadOR':"operador", 'flight_no':"numero_de_vuelo", 'route':"tipo_de_ruta",
       'ac_type':"tipo_de_aeronave", 'registration': "matricula", 'cn_ln': "numero_construccion_ldm", 'all_aboard': "total_a_bordo", 'PASAJEROS A BORDO': "pasajeros_a_bordo",
       'crew_aboard':"tripulacion_a_bordo", 'cantidad de fallecidos': "total_fallecidos", 'passenger_fatalities': "pasajeros_fallecidos",
       'crew_fatalities': "tripulacion_fallecidos", 'ground':"suelo", 'summary':"resumen"}, inplace= True)
df
```

```python
# como los valores nulos de nuestro dataset estan representados con un signo de interrogacion los cambiamos a valores de tipo NaN
df.replace("?", np.nan, inplace = True)
df.isnull().sum()
```

```python
# Ahora como primer paso la columna que a priori mas modificaciones debe tener es la columna "hora declarada"
df["hora_declarada"].sample(20)
# Asi que imprimimos un sample de esta columna para ver que valores deben de modificarse
# Para que poco a poco todos queden con el mismo formato de tipo "##:##"
```

```python
# Como primer cambio realizamos el reemplazo de las letras "c" sueltas en nuestro df que no proporcionan ninguna info importante 
# Y ademas entorpecen nuestro labor
df["hora_declarada"] = df["hora_declarada"].str.replace ("c", "", regex= True)
```

```python
# Procedemos a imprimir otro sample para verificar otros tipos de modificaciones que debemos realizar
df["hora_declarada"].sample(20)
```

```python
# Como tenemos valores con ":" fuera de lugar los vamos a eliminar
df["hora_declarada"] = df["hora_declarada"].str.replace (":", "", regex= True)
```

```python
df["hora_declarada"].sample(20)
```

```python
# Para este punto en principio nuestros datos deberian ser en su totalidad de una longitud igual a 4
df[df["hora_declarada"].str.len()>4 ]
```

```python
# Podemos notar en este caso que tenemos aun valores por corregir y proseguimos con esto
df["hora_declarada"] = df["hora_declarada"].str.replace (";", "", regex= True)
df["hora_declarada"] = df["hora_declarada"].str.replace (" ", "", regex= True)
df["hora_declarada"] = df["hora_declarada"].str.replace ("Z", "", regex= True)
```

```python
# volvemos a verificar que nuestros datos tengan maximo una longitud de 4
df[df["hora_declarada"].str.len()>4 ]
```

```python
# Imprimimos nuestro dataframe completo para seguir evaluando posibles modificaciones
df
```

```python
# Sacamos la info de nuestro dataframe para ver que tipo de datos tenemos
df.info()
```

```python
# como ya tenemos la columna hora lista para ser convertida a un formato tiempo seguimos con ese paso:
df["hora_declarada"]=pd.to_datetime(df["hora_declarada"], format='%H%M').dt.time
df
```

```python
# Ademas de la columna de "hora_declarada" tenemos esta columna "fecha" que tambien debe ser convertida a un formato de tiempo
df['fecha'] = pd.to_datetime(df['fecha'], format='%B %d, %Y', errors='coerce')
df
```

```python
# Como logramos convertir de manera exitosa nuestra columna "fecha" el siguiente paso es verificar si hubieron fallos
# durante el proceso
df.info()
```

```python
# Reacomodamos las columnas de nuestro dataframe para tener un orden logico en ellas
nuevo_orden_columnas = ["fecha", "hora_declarada", "ruta", "operador", "numero_de_vuelo", "tipo_de_ruta", "tipo_de_aeronave", "matricula", "numero_construccion_ldm", "pasajeros_a_bordo", "tripulacion_a_bordo", "total_a_bordo", "pasajeros_fallecidos", "tripulacion_fallecidos", "total_fallecidos", "suelo", "resumen"]
df = df.reindex(columns=nuevo_orden_columnas)
# Ademas de eso dropeamos las columnas que consideremos inutiles
df.drop(columns=["numero_de_vuelo"])
# E imprimimos nuevamente el dataframe
```

```python
# Por ultimo corregimos el tipo de dato en nuestras columnas numericas
columnas_convertir = ["pasajeros_a_bordo", "tripulacion_a_bordo", "total_a_bordo",
                      "pasajeros_fallecidos", "tripulacion_fallecidos", "total_fallecidos",
                      "suelo"]

# Aplicar la conversión a las columnas seleccionadas
df[columnas_convertir] = df[columnas_convertir].apply(pd.to_numeric, errors='coerce', downcast='integer')

df.info()
```

```python
# Exportamos los datos ya normalizados
df.to_csv("df.csv", index= False)
```
